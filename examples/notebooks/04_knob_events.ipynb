# 旋钮交互：转出一个表盘

收集旋钮事件，把转动量累积成指针角度画出来。**运行采集单元格时请转动眼镜的旋钮**（两个方向都试试）。

```python
from edu_notebook import EduSession

s = EduSession()
```

```python
print("开始转旋钮！采集 15 秒……")
events = s.events(15)
knob = [d for kind, detail in events if kind == "KNOB"
        for d in [EduSession.knob_delta(detail)] if d]
print("收到 %d 个事件，其中旋钮 %d 个" % (len(events), len(knob)))
if not knob:
    print("没收到旋钮事件——确认转的是旋钮、连接未断，然后重跑本单元格")
```

```python
import math
import matplotlib.pyplot as plt

angle = 0.0
trace = [0.0]
for _direction, dx in knob:
    angle += dx * 6      # 每格 6 度
    trace.append(angle)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))
ax1.plot(trace); ax1.set_title("accumulated angle"); ax1.set_xlabel("event #"); ax1.grid(True)
rad = math.radians(trace[-1] - 90)
ax2.add_artist(plt.Circle((0, 0), 1.0, fill=False, linewidth=2))
ax2.plot([0, math.cos(rad)], [0, -math.sin(rad)], linewidth=3)
ax2.set_xlim(-1.2, 1.2); ax2.set_ylim(-1.2, 1.2)
ax2.set_aspect("equal"); ax2.axis("off"); ax2.set_title("dial: %.0f°" % trace[-1])
plt.show()
```
