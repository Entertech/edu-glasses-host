# 拍照与图像处理

用眼镜拍一张照片，然后用 PIL/matplotlib 做基础图像分析。

前置：已在系统蓝牙配对并连接眼镜（`EDU-Glasses-xxxx`）。依赖：`pip install matplotlib pillow`。

```python
from edu_notebook import EduSession

s = EduSession()   # macOS 自动发现；其他平台 EduSession(bt="AA:BB:CC:DD:EE:FF")
```

```python
path = s.photo("notebook_shot.jpg")   # 首拍含相机唤醒，最长约 25 秒
print(path)
```

```python
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open(path)
print("尺寸:", img.size, " 模式:", img.mode)
plt.figure(figsize=(6, 8))
plt.imshow(img)
plt.axis("off")
plt.title("EDU glasses photo")
plt.show()
```

```python
# RGB 直方图：光照实验的第一课
import numpy as np

arr = np.asarray(img)
plt.figure(figsize=(8, 3))
for i, color in enumerate("rgb"):
    hist, bins = np.histogram(arr[..., i], bins=64, range=(0, 255))
    plt.plot(bins[:-1], hist, color=color, label=color.upper())
plt.legend(); plt.title("RGB histogram"); plt.xlabel("pixel value")
plt.show()
```

```python
# 灰度 + 边缘检测
from PIL import ImageFilter

edges = img.convert("L").filter(ImageFilter.FIND_EDGES)
plt.figure(figsize=(6, 8))
plt.imshow(edges, cmap="gray")
plt.axis("off")
plt.title("edges")
plt.show()
```
