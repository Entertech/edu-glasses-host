# 光敏传感器实时曲线

每秒读一次 ALS（环境光原始计数），实时画曲线。**运行期间用手遮住眼镜的光敏窗口再放开**，观察曲线变化。

ALS 是原始 16-bit 计数值（不是 lux），适合做相对比较。

```python
from edu_notebook import EduSession

s = EduSession()
```

```python
import time
import matplotlib.pyplot as plt
from IPython.display import clear_output

SAMPLES = 20        # 采样次数（约 1 次/秒）
values, stamps = [], []
t0 = time.time()
for i in range(SAMPLES):
    reading = s.sensors()
    values.append(reading["als"])
    stamps.append(time.time() - t0)
    clear_output(wait=True)
    plt.figure(figsize=(8, 3))
    plt.plot(stamps, values, marker="o")
    plt.xlabel("s"); plt.ylabel("ALS raw")
    plt.title("ALS live (%d/%d)  当前=%d" % (i + 1, SAMPLES, values[-1]))
    plt.grid(True)
    plt.show()
    time.sleep(1)
```

```python
print("最小 %d / 最大 %d / 变化幅度 %d" % (min(values), max(values), max(values) - min(values)))
```
