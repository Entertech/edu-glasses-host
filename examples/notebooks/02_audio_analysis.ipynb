# 录音与音频分析

用眼镜麦克风录几秒声音，画波形和声谱图。

提示：录音时对眼镜说话或弹个响指，图会好看很多。WAV 输出需要 opuslib + libopus（见仓库 README）。

```python
from edu_notebook import EduSession

s = EduSession()
```

```python
path = s.record("notebook_voice.wav", seconds=5)
print(path)
```

```python
import wave
import numpy as np

with wave.open(str(path)) as w:
    rate = w.getframerate()
    pcm = np.frombuffer(w.readframes(w.getnframes()), dtype=np.int16)
print("采样率 %d Hz，时长 %.1f s" % (rate, len(pcm) / rate))
```

```python
import matplotlib.pyplot as plt

t = np.arange(len(pcm)) / rate
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(9, 6))
ax1.plot(t, pcm, linewidth=0.4)
ax1.set_title("waveform"); ax1.set_xlabel("s")
ax2.specgram(pcm, Fs=rate, NFFT=512, noverlap=256)
ax2.set_title("spectrogram"); ax2.set_xlabel("s"); ax2.set_ylabel("Hz")
plt.tight_layout(); plt.show()
```
